### Question 14
What code would you use to set up a view of all movies that did not get released?


```python
%%sql 

CREATE VIEW Not_Released AS 

SELECT 
    * 
FROM 
    movies 
WHERE 
    release_status <> 'Released'
```

### Question 17
What is the most popular movie that was made after 01/01/2000 with a budget of more than $100 000 000? (Hint: Use the popularity field in the Movies table. Larger numbers are more popular.)


```python
%%sql 
SELECT 
    title,
    popularity
FROM 
    movies
WHERE 
    release_date > '2000-01-01'
    AND budget > 100000000
ORDER BY 
    popularity DESC
LIMIT 5;
```

### Question 19
How many movies in the database were produced by Pixar Animation Studios?



```python
%%sql 

SELECT 
    COUNT(*)
FROM 
    movies m
JOIN 
    ProductionCompanyMap pcm 
    ON m.movie_id = pcm.movie_id
JOIN 
    ProductionCompanies pc 
    ON pcm.production_company_id = pc.production_company_id
WHERE 
    pc.production_company_name = 'Pixar Animation Studios';
```

### Question 20
How many movies are in the database that are both a Romance and a Comedy?


```python
%%sql 

SELECT 
    COUNT(DISTINCT m.movie_id) AS number_of_movies
FROM 
    movies m
JOIN 
    GenreMap gm1 
    ON m.movie_id = gm1.movie_id
JOIN 
    Genres g1 
    ON gm1.genre_id = g1.genre_id
JOIN 
    GenreMap gm2 
    ON m.movie_id = gm2.movie_id
JOIN 
    Genres g2 
    ON gm2.genre_id = g2.genre_id
WHERE 
    g1.genre_name = 'Romance' 
    AND g2.genre_name = 'Comedy'
```

### Question 21
What is the most popular action movie that has some German in it? (Hint: The German word for German is Deutsch)

```python
%%sql 

SELECT 
    m.title, 
    m.popularity
FROM 
    Movies m
JOIN 
    LanguageMap lm 
    ON m.movie_id = lm.movie_id
JOIN 
    Languages l 
    ON lm.iso_639_1 = l.iso_639_1
JOIN 
    GenreMap gm 
    ON m.movie_id = gm.movie_id
JOIN 
    Genres g 
    ON gm.genre_id = g.genre_id
WHERE 
    g.genre_name = 'Action' 
    AND l.language_name = 'Deutsch'
ORDER BY 
    m.popularity DESC
LIMIT 1;
```

### Question 22
In how many movies did Tom Cruise portray the character Ethan Hunt?



```python
%%sql 

SELECT 
    COUNT(*)
FROM 
    Movies m
JOIN 
    Casts c 
    ON m.movie_id = c.movie_id
JOIN 
    Actors a 
    ON c.actor_id = a.actor_id
WHERE 
    a.actor_name = 'Tom Cruise' 
    AND c.characters = 'Ethan Hunt'
```

### Question 24
How many movies contain at least one of the official South African Languages, Afrikaans or Zulu?

```python
%%sql 

SELECT 
    COUNT(DISTINCT m.movie_id)
FROM 
    Movies m
JOIN 
    LanguageMap lm 
    ON m.movie_id = lm.movie_id
JOIN 
    Languages l 
    ON lm.iso_639_1 = l.iso_639_1
WHERE 
   l.language_name LIKE '%Afrikaans%'
    OR  l.language_name LIKE '%Zulu%'
```

```python
%%sql 

SELECT 
   *
FROM 
    languages
```
